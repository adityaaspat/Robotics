```python
import json
import numpy as np
import cv2
from scipy.optimize import curve_fit
import os

# Define file paths
original_file_path = '/media/aditya/ExtraSpace/track_images/track_images.json'
mirrored_file_path = '/media/aditya/ExtraSpace/track_images/mirrored_track_images.json'

# Define the log function for curve fitting
def log_func(x, a, b, c):
    return a * np.log(b * (x + 1e-6)) + c

# Initialize the data dictionary
data = {}

# Function to load JSON data into the data dictionary
def load_data(json_file_path):
    with open(json_file_path, 'r') as file:
        json_data = json.load(file)
    for image_path, points in json_data.items():
        data[image_path] = points

# Load original and mirrored JSON data into the same dictionary
load_data(original_file_path)
load_data(mirrored_file_path)

# Iterate through the entries and extract points
for image_path, points in data.items():
    print(f"Processing Image: {image_path}")
    points = np.array(points)

    # Extract x and y coordinates, adjusting for image size
    X = points[:, 1].reshape(-1, 1) * 224 / 800  # y-coordinates adjusted for height
    y = points[:, 0] * 224 / 800                 # x-coordinates adjusted for width

    # Fit the curve using the log function
    try:
        params, _ = curve_fit(log_func, X.reshape(-1), y)
        a, b, c = params

        # Generate fitted points
        y_fitted = np.linspace(75, 175, 100, dtype=int)
        x_fitted = log_func(y_fitted, a, b, c)
        fitted_points = np.column_stack((x_fitted, y_fitted))

        # Update the dictionary with fitted points
        data[image_path] = fitted_points.tolist()
    except Exception as e:
        print(f"Curve fitting failed for {image_path}: {e}")
        data[image_path] = points.tolist()  # Store original points if fitting fails

print("Data processing completed.")
```

```python
class PolynomialRegressionModel(nn.Module):
    def __init__(self):
        super(PolynomialRegressionModel, self).__init__()
        
        efficient_net = efficientnet_b0()
        
        # remove the classifier part and keep the feature extractor
        self.feature_extractor = nn.Sequential(*list(efficient_net.children())[:-1])
        
        #flatten the output of EfficientNet feature extractor
        self.flatten = nn.Flatten()
        
        # define  MLP 
        self.net = nn.Sequential(
            nn.Linear(efficient_net.classifier[1].in_features, 200),
            nn.ReLU(),
            
        )

    
    def forward(self, x):
        # Pass through EfficientNet feature extractor
        features = self.feature_extractor(x)
        
        # Flatten the features
        features = self.flatten(features)
        
        output = self.net(features)
        
        return output
 
# Instantiate and check the model structure
model = PolynomialRegressionModel()
print(model)
```

```python

class LaneDataset(Dataset):
    def __init__(self, image_paths, lane_points, transform=None):
        self.image_paths = image_paths
        self.lane_points = lane_points
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image = Image.open(self.image_paths[idx]).convert('RGB')
      
        points=torch.tensor(self.lane_points[idx],dtype=torch.float32)
        if self.transform:
            image = self.transform(image)

        return image, points

my_transform = transforms.Compose([
    transforms.Resize(size=(224,224)),
    transforms.ToTensor(),          
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Normalize along each channel
])
batch_size=32

train_dataset = LaneDataset(train_image_paths, train_lane_points, transform=my_transform)
test_dataset = LaneDataset(test_image_paths, test_lane_points, transform=my_transform)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

train_dataset.__getitem__(78)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(torch.cuda.is_available())
model = model.to(device)
```

```python
criterion = MSELoss()

optimizer = optim.AdamW(model.parameters(), lr=1e-6)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer,T_max=10)
output_size = 100
```

```python

num_epochs = 200  # 

for epoch in range(num_epochs):
    model.train() 
    running_loss = 0.0
    

    for images, points in train_loader:
        images = images.to(device)
        points = points.to(device)
        
        optimizer.zero_grad()
        
        outputs = model(images)
        outputs = outputs.view(-1,output_size,2)

        loss = criterion(outputs, points)
        
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
    
    scheduler.step()

    
    model.eval()  
    test_loss = 0.0
    


    with torch.no_grad(): 
        for images, points in test_loader:
            images = images.to(device)
            points = points.to(device)
            
            # Forward pass
            outputs = model(images)
            outputs =  outputs.view(-1,output_size,2)
            
            
            
            loss = criterion(outputs, points)
            
            # Accumulate test loss
            test_loss += loss.item()

    

    # Print statistics for training and testing
    print(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {running_loss/len(train_loader):.4f},Test Loss: {test_loss/len(test_loader):.4f}")
```

```python
import cv2
import glob
import matplotlib.pyplot as plt
from PIL import Image
import torch
from scipy.optimize import curve_fit
import numpy as np

def log_func(x, a, b, c):
    return a * np.log(b * x) + c

# Initialize variables for video creation
frame_size = (224, 224)  # Assuming resized images are 224x224
output_video_path = "output_video.avi"
frame_rate = 10  # Frames per second
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # Codec for AVI file
video_writer = cv2.VideoWriter(output_video_path, fourcc, frame_rate, frame_size)

model.eval()

def plot_test_points_on_image(image_path):
    # Read the image
    test_path = image_path
    test_image = Image.open(test_path).convert('RGB')
    image_in = test_image.copy()
    # Transform the image (same transformation as during training)
    image_in = my_transform(image_in)  # Apply the same transformation
    with torch.no_grad():  # No gradient calculation during testing
        image_in = image_in.to(device)
        if image_in.dim() == 3:
            image_in = image_in.unsqueeze(0) 
        outputs = model(image_in)
        outputs = outputs.cpu().numpy()[0]

    test_image = cv2.imread(test_path)
    # Convert BGR to RGB (OpenCV uses BGR by default)
    img_rgb = cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB)
    img_rgb = cv2.resize(img_rgb, (224, 224))
    const = 1  # Adjust scale if necessary
    outputs = outputs.reshape(-1, 2)
    X = outputs[:, 1].reshape(-1, 1)  # Extract y-coordinates and calculate x
    Y = outputs[:, 0]
    params, _ = curve_fit(log_func, X.reshape(-1), Y)
    a, b, c = params
    y = np.linspace(100, 224, 100, dtype=int)
    x = log_func(y, a, b, c)
    for i in range(len(outputs)):   
        cv2.circle(img_rgb, (int(x[i] * const), int(y[i] * const)), radius=2, color=(255, 0, 0), thickness=-1)

    # Convert RGB to BGR (OpenCV requires BGR format for VideoWriter)
    img_bgr = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR)
    video_writer.write(img_bgr)  # Add frame to video

# Process each image and add to the video
image_files = glob.glob("/media/aditya/ExtraSpace/test_images/*.jpg")
image_files.sort()  # Sort alphabetically (or numerically if filenames allow)

for image_path in image_files:
    plot_test_points_on_image(image_path)

# Release the video writer
video_writer.release()
print(f"Video saved to {output_video_path}")
```
